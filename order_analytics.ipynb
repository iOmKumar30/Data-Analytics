```python
# extract the data from the downloaded file using the zip command
import zipfile
zip_path = 'orders.csv.zip'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('.')
```

```python
# filling in the NULL values
import pandas as pd
import numpy as np
data = pd.read_csv('orders.csv')
data.head(10)
# as we can see there are "not available" values in the "ship mode" column and we need to assign them as null values
data['Ship Mode'].unique()
data['Ship Mode'] = data['Ship Mode'].replace(['Not Available', 'unknown'], np.nan)
data.head(10)
```

```python
# renaming the column names to conventional format i.e. Order Id : order_id
columns = data.columns
columns = [allColumn.lower().replace(' ', '_') for allColumn in columns]
data.columns = columns
data.head(5)
```

```python
# we have cost price, list price(MRP) and discount percent so now we need to create discount column, sale price column and profit column
data['discount'] = data['list_price'] * (data['discount_percent'] / 100) 
data['sale_price'] = data['list_price'] - data['discount']
data['profit'] = data['sale_price'] - data['cost_price']
data.head(5)
```

```python
data.dtypes
```

```python
# As we saw the order_data is there as an object so its ideal to convert it into a date format
data['order_date'] = pd.to_datetime(data['order_date'], format='%Y-%m-%d')
data.dtypes
```

```python
# drop the unnecessary columns
data.drop(columns=['cost_price', 'discount_percent', 'list_price'], inplace=True)
data.head(5)
```

```python
from sqlalchemy import create_engine

username = 'postgres'
password = 'postgres'
database = 'order_analytics'  
host = 'localhost'
port = '5432'

connection_string = f'postgresql://{username}:{password}@{host}:{port}/{database}'

engine = create_engine(connection_string)

try:
    with engine.connect() as connection:
        data.to_sql(name='orders', con=connection, if_exists='append', index=False)
    print("Data inserted successfully.")
except Exception as e:
    print("An error occurred:", e)
```
